```python
import s3fs
import os
import datetime as dt
import xarray as xr
```

```python
bucket = 's3://noaa-gefs-pds'
local_target = './gefs-data'
date_range = ('2023-07-13', '2023-07-15')
```

```python
bucket_paths=[]
files=[]

date = '20231105'
start_hr = '00'
members = ('gec00','gep01')
model_hrs = ('000','006','012','018')
#0p50 is 0.50 degree resolution
#Build a list of target file names
for m in members:
    for h in model_hrs:
        file = m+'.t'+start_hr+'z.pgrb2a.0p50.f'+h
        files.append(file)
        bucket_paths.append(bucket+'/gefs.'+date+'/'+start_hr+'/atmos/pgrb2ap5/'+file)

#Copy files from the s3 bucket to a local folder
if not os.path.exists(local_target):
    os.mkdir(local_target)
s3 = s3fs.S3FileSystem(anon=True)
s3.get(bucket_paths, local_target)
print(files)
```
